##03 L04 Contractor
[Contractor] ผู้รับเหมาก่อสร้าง
ผู้รับเหมาก่อสร้างได้รับงานสร้างรูปปั้นเชิดชูเกียรติ จึงได้ติดต่อจ้างน้องๆ CPE 36 ให้ทำโปรแกรมรับคำสั่งจากลูกค้าว่าจะสร้างเป็นรูปทรงอะไร ขนาดเท่าไหร่ จากนั้นทำการคำนวณว่าจะต้องใช้ปริมาณวัสดุเท่าใด ราคาวัสดุทั้งหมดเท่าไหร่สำหรับงานนี้

ภายในโปรแกรมต้องมีฟังก์ชันดังต่อไปนี้

inputShape() สำหรับรับค่า และ คืนค่าข้อมูลนำเข้าที่รับมา เพียงค่าเดียวซึ่งเป็นตัวเลขระบุว่าต้องสร้างเป็นรูปทรงอะไร โดยข้อมูลนำเข้าประกอบด้วยหมายเลข 1 (ทรงกลม) 2 (ทรงกรวย) 3 (ทรงกระบอก) 4 (ปริซึมฐานสี่เหลี่ยม) 5 (พีระมิดฐานสี่เหลี่ยม) เพื่อนำไปหาต่อว่าต้องคำนวณด้วยฟังก์ชันใด
calculateSphere() สำหรับรับค่ารัศมี คำนวณ แสดงผล และ คืนค่าปริมาตรทรงกลม
calculateCone() สำหรับรับค่ารัศมี,ความสูง คำนวณ แสดงผล และ คืนค่าปริมาตรทรงกรวย
calculateCylinder() สำหรับรับค่ารัศมี,ความสูง คำนวณ แสดงผล และ คืนค่าปริมาตรทรงกระบอก
calculatePrism() สำหรับรับค่าความกว้าง,ยาว,สูง คำนวณ แสดงผล และ คืนค่าปริมาตรปริซึมฐานสี่เหลี่ยม (Rectangular prisms)
calculatePyramid() สำหรับรับค่าความกว้าง,ยาว,สูง คำนวณ แสดงผล และ คืนค่าปริมาตรพีระมิดฐานสี่เหลี่ยม (Rectangle-based pyramids) ให้ใช้สูตร กว้าง * ยาว * สูง / 3 เท่านั้น ห้ามใช้ (1/3) * กว้าง * ยาว * สูง
calculatePrice(v) สำหรับรับค่าราคาต่อ 1 ลบ.ม. คำนวณและแสดงผลราคาของวัสดุ
สามารถดูส่วนประกอบต่างๆภายในฟังก์ชันดังตัวอย่าง

กำหนดให้ค่า pi = 3.141592653589793

ให้ใช้ตัวแปร Input เป็น int ทุกตัว

ข้อมูลส่งออกทุกค่าเป็นทศนิยม 2 ตำแหน่ง

สูตรการหาปริมาตรรูปทรงต่าง ๆ (คลิ๊ก)
```
Example 1
Input Shape: 1
Input radius(meter): 5
The volume is 523.60 cubic meter.
Price(Bath) per 1 cubic meter: 2
The price is 1047.20 Baht.
Example 2
Input Shape: 2
Input radius(meter): 3
Input height(meter): 4
The volume is 37.70 cubic meter.
Price(Bath) per 1 cubic meter: 5
The price is 188.50 Baht.
Example 3
Input Shape: 3
Input radius(meter): 4
Input height(meter): 5
The volume is 251.33 cubic meter.
Price(Bath) per 1 cubic meter: 6
The price is 1507.96 Baht.
Example 4
Input Shape: 4
Input width(meter): 4
Input length(meter): 2
Input height(meter): 2
The volume is 16.00 cubic meter.
Price(Bath) per 1 cubic meter: 10
The price is 160.00 Baht.
Example 5
Input Shape: 5
Input width(meter): 6
Input length(meter): 3
Input height(meter): 4
The volume is 24.00 cubic meter.
Price(Bath) per 1 cubic meter: 2
The price is 48.00 Baht.
```

````python
import math

def inputShape():
    return(int(input("Input Shape: ")))
    
def calculateSphere():
    r = int(input("Input radius(meter): "))
    v = 4/3*math.pi*r*r*r
    print(f"The volume is {v:.2f} cubic meter.")
    return v

def calculateCone():
    r = int(input("Input radius(meter): "))
    h = int(input("Input height(meter): "))
    v = math.pi*r*r*h/3
    print(f"The volume is {v:.2f} cubic meter.")
    return v

def calculateCylinder():
    r = int(input("Input radius(meter): "))
    h = int(input("Input height(meter): "))
    v = math.pi*r*r*h
    print(f"The volume is {v:.2f} cubic meter.")
    return v

def calculatePrism():
    w = int(input("Input width(meter): "))
    l = int(input("Input length(meter): "))
    h = int(input("Input height(meter): "))
    v = w*l*h
    print(f"The volume is {v:.2f} cubic meter.")
    return v

def calculatePyramid():
    w = int(input("Input width(meter): "))
    l = int(input("Input length(meter): "))
    h = int(input("Input height(meter): "))
    v = w*l*h/3
    print(f"The volume is {v:.2f} cubic meter.")
    return v

def calculatePrice(v):
    price = int(input("Price(Bath) per 1 cubic meter: "))
    print(f"The price is {price*v:.2f} Baht.")

shape = inputShape()
if shape == 1: v = calculateSphere()
elif shape == 2: v = calculateCone()
elif shape == 3: v = calculateCylinder()
elif shape == 4: v = calculatePrism()
elif shape == 5: v = calculatePyramid()
calculatePrice(v)
````

##04 L04 our_year
our_year ปีนี้มาแน่
การแข่งขันฟุตบอลพรีเมียร์ลีกเป็นการแข่งกันแบบเจอกันหมด โดยเงื่อนไขในการเป็นแชมป์ฟุตบอลพรีเมียร์ลีก คือ
     ทีมที่มีแต้มมากที่สุดจะเป็นแชมป์
     หากมีทีมที่มีแต้มมากที่สุดเท่ากัน ทีมที่มีจำนวนประตูได้/เสีย (Goal difference) มากกว่าจะเป็นแชมป์
     หากมีแต้ม และ จำนวนประตูได้เสียเท่ากัน ต้องแข่งรอบ play-off และทีมที่ได้คะแนนมากกว่าในรอบนี้จะเป็นแชมป์

ในการแข่งขันฟุตบอลพรีเมียร์ลีกปีนี้ได้ดำเนินมาถึงแมทช์สุดท้ายของซีซั่นโดยที่ทีมลิเวอร์พูล และ แมนเชสเตอร์ซิตี้ มีแต้ม และ จำนวนประตูได้/เสีย เท่ากัน จึงทำให้ผลการตัดสินแชมป์ขึ้นอยู่กับแมทช์สุดท้าย (ไม่ใช่รอบ play-off) โดยในแมทช์สุดท้ายนี้
     ลิเวอร์พูลต้องเจอกับคู่ปรับตลอดกาล แมนเชสเตอร์ยูไนเต็ด
     แมนเชสเตอร์ซิตี้ต้องพบกับทีมท้ายตาราง นิวคาสเซิล
จงเขียนโปรแกรมเพื่อตัดสินว่าทีมใดจะได้เป็นแชมป์พรีเมียร์ลีก (สมมติว่าผลการแข่งขันมีเพียง แพ้ หรือ ชนะ เท่านั้น)
```
Example1
What's the result of Manchester city's match? won
What's the result of Liverpool's match? lost
Manchester city is the champion of Premier League.
Note: เนื่องจาก Manchester city ชนะ จึงทาให้ Manchester city เป็นแชมป์โดยปริยาย ในทางกลับกันหาก Liverpool ชนะ ก็จะทาให้ Liverpool เป็นแชมป์โดยปริยาย
Example2
What's the result of Manchester city's match? won
What's the result of Liverpool's match? won
What is the margin that Manchester city won by? 3
What is the margin that Liverpool won by? 1
Manchester city is the champion of Premier League.
Note: หากทั้งคู่ชนะในการแข่งขันแมทช์สุดท้าย ให้ดูที่จานวนประตู ทีมไหนที่ทาประตูได้มากกว่าก็จะเป็นแชมป์
Example3
What's the result of Manchester city's match? won
What's the result of Liverpool's match? won
What is the margin that Manchester city won by? 4
What is the margin that Liverpool won by? 4
Which team won the play-off match?(Manchester city/Liverpool) Manchester city
Manchester city is the champion of Premier League.
```
Note: หากทั้งคู่ชนะในการแข่งขันแมทช์สุดท้าย และ ได้จานวนประตูในแมทช์สุดท้ายเท่ากัน ให้แข่งรอบ play-off ฝ่ายที่ชนะก็จะเป็นแชมป์

````python
def check_won(man,liver):
    if man == liver: return 0
    elif man and not liver:
        print("Manchester city is the champion of Premier League.")
        return 1
    else:
        print("Liverpool is the champion of Premier League.")
        return 1

man= 1 if input("What's the result of Manchester city's match? ") == "won" else 0
liver = 1 if input("What's the result of Liverpool's match? ") == "won" else 0
if not check_won(man,liver):
    man_s = int(input("What is the margin that Manchester city won by? "))
    liver_s = int(input("What is the margin that Liverpool won by? "))
    if man_s>liver_s:
        man = 1
        liver = 0
    elif liver_s>man_s:
        liver=1
        man=0
    if not check_won(man, liver):
        winner = input("Which team won the play-off match?(Manchester city/Liverpool) ")
        print(f"{winner} is the champion of Premier League.")
````

จงเขียนฟังก์ชันชื่อ check_prime(n) โดยรับค่า n จากโปรแกรมหลักเพื่อตรวจสอบว่าค่า n นั้นเป็นจำนวนเฉพาะหรือไม่

ให้ฟังก์ชันดังกล่าวคืนค่าเป็นบูล (นั่นคือ คืนค่า True ถ้าเป็นจำนวนเฉพาะ และคืนค่า False ถ้าไม่ใช่)

ตัวอย่างการเรียกใช้งานผ่านโปรแกรมหลักแสดงด้านล่าง
```
Example 1
Enter number: 100
100 is not a prime number.
Example 2
Enter number: 139
139 is a prime number.
```

```
ตัวอย่างของโปรแกรมหลักที่เรียกใช้ฟังก์ชันดังกล่าวคือ

n = int(input("Enter number: "))
if check_prime(n):
    print(n,"is a prime number.")
else:
    print(n,"is not a prime number.")
```

````python
def check_prime(n):
    arr = [0]*(n+100)
    # print(arr)
    arr[1] = 1
    for i in range(2,n+1):
        if arr[i] == 0:
            # print(f"{i} ")
            for j in range(i*i,n+1,i):
                arr[j] = 1
    if not arr[n]: return True
    else: return False
````

ตัวประกอบของ 70 มี 8 ตัว คือ 1, 2, 5, 7, 10, 14, 35 และ 70

```
ส่วนของโปรแกรมหลักที่เรียกใช้ฟังก์ชันดังกล่าว เป็นดังด้านล่าง

n = int(input("Enter n: "))
c = factor_count(n)
print(c)
```



````python
def factor_count(n):
    # l_factor = []
    counter=0
    for i in range(1,int(n**0.5)+1):
        if n%i == 0:
            # l_factor.append(i)
            counter+=1
            if n//i != i:
                # l_factor.append(n//i)
                counter+=1
    # print(l_factor)
    return counter
````

เขียนโปรแกรมเพื่อหาค่าฟิโบนัชชีในลำดับที่ n เมื่อ n เป็นจำนวนเต็มบวก

หมายเหตุ ถ้าใช้วิธีการแบบ recursive จะไม่ผ่าน 2 cases สุดท้าย เพราะเวลาที่ใช้จะมากเกินกว่าเวลาที่ยอมรับได้
```
ตัวอย่างโปรแกรม 1

ตัวอย่างโปรแกรม 2
Enter n: 8
fibo(8) = 21
```

````python
def fibo(n):
    fibo_l = [0]*100000
    # print(fibo_l[10])
    fibo_l[1] = 1 
    fibo_l[2] = 1
    for i in range(2+1,n+1):
        # print(i)
        fibo_l[i] = fibo_l[i-1] + fibo_l[i-2]
        # print(fibo_l)
    return fibo_l[n]

n = int(input("Enter n: "))
print("fibo({a}) = {b}".format (a=n, b=fibo(n)))
````

##19 การขยายพันธุ์ของแบคทีเรีย
การขยายพันธุ์ของแบคทีเรีย
ในการศึกษาวิจัยการขยายพันธุ์ของแบคทีเรียสายพันธุ์ bal-x บนอาหาร พบว่า ในแต่ละวันจะมีแบคทีเรียสายพันธุ์ bal-x จะขยายจำนวนเพิ่มขึ้นจากเดิมวันละ d% จากเดิม และทำปฏิกิริยากับแบคทีเรียอื่นในอากาศเพิ่มอีก aug ตัว

เช่น หากเริ่มต้นมีแบคทีเรีย bal-e (หนึ่งในสายพันธุ์ของ bal-x) อยู่ 1000 ตัว มีอัตราการขยายตัววันละ 2.0% และทำปฏิกิริยากับแบคทีเรียอื่นเพิ่มขึ้นวันละ 30 ตัว
เมื่อผ่านไป 1 วัน แบคทีเรีย bal-e จะมีจำนวนเพิ่มขึ้นเป็น 1000 + 1000 * 0.02 + 30 = 1050 ตัว
เมื่อผ่านไป 2 วัน แบคทีเรีย bal-e จะมีจำนวนเพิ่มขึ้นเป็น 1050 + 1050 * 0.02 + 30 = 1101 ตัว
เมื่อผ่านไป 3 วัน แบคทีเรีย bal-e จะมีจำนวนเพิ่มขึ้นเป็น 1101 + 1101 * 0.02 + 30 = 1153 ตัว (จำนวนแบคทีเรียเป็นจำนวนเต็มเสมอ)

นักวิจัยมีข้อมูลจำนวนแบคทีเรียเริ่มต้น (p0) อัตราการขยายตัวของแบคทีเรียต่อวัน (percent)% จำนวนแบคทีเรียที่เพิ่มขึ้นเมื่อทำปฏิกิริยากับแบคทีเรียในอากาศ (aug จำนวนอาจลดลง หรือเป็น 0 ก็ได้)

นักวิจัยต้องการทราบว่า จะต้องใช้จำนวนกี่วันนับจากวันแรกที่แบคทีเรียสายพันธุ์นี้ จะมีจำนวนมากกว่าหรือเท่ากับ p

ในข้อนี้ ให้เขียนฟังก์ชัน nb_year(p0, percent, aug, p) เพื่อคืนค่าจำนวนวัน นับจากวันแรกที่แบคทีเรียสายพันธุ์นี้ขยายพันธุ์จนมีจำนวนมากกว่าหรือเท่ากับ p
```
ตัวอย่างการเรียกฟังก์ชัน
print( nb_year(1000, 2, 30, 1150) )

3
print( nb_year(1500000, 0.25, 1000, 2000000) )

94
```
ส่งคำตอบเฉพาะ Function Definition ของ nb_year(p0, percent, aug, p)

````python
def nb_year(p0, percent, aug, p):
    counter = 0
    while p0 < p:

        p0 = int(p0 + p0*percent/100 + aug)
        # print(p0)
        counter += 1
    return counter
````

##20 L04 Sale
Sale
ร้านค้าแห่งหนึ่งจัดโปรโมชั่นลดราคาสินค้าที่ซื้อในแต่ละวัน โดยที่เริ่มต้นวันแรกที่มาซื้อจะได้ลดราคา5% และถ้ามาซื้อของที่ร้านค้านี้ติดต่อกัน จะได้ลดราคาจากเดิม1%ทุกครั้ง

ให้น้องๆเขียนโปรแกรมรับจำนวนวันที่ลูกค้าซื้อติดกันและราคาสินค้าในแต่ละวัน แล้วนำมาคิดว่าลูกค้าคนนี้ได้ทำการจ่ายเงินให้กับร้านค้าเป็นจำนวนเท่าไหร่แล้ว
```
Example 1
How many day : 2
Input price in day 1 : 100
Input price in day 2 : 100
Summary price = 189.00
Example 2
How many day : 5
Input price in day 1 : 123.5
Input price in day 2 : 120.65
Input price in day 3 : 210.1
Input price in day 4 : 45
Input price in day 5 : 52
Summary price = 514.85
Example 3
How many day : 4
Input price in day 1 : 22 
Input price in day 2 : 23
Input price in day 3 : 24
Input price in day 4 : 25
Summary price = 87.84
```

````python
day = int(input("How many day : "))
sum = 0
discount = 5
for i in range(day):
    cash = float(input(f"Input price in day {i+1} : "))
    sum += cash*(100-discount)/100
    discount+=1
print(f"Summary price = {sum:.2f}")
````

````python
import math

class MyMath:
  a = 3
  for i in range(1,51):
    if i%2 != 0:
      a += 4/(i*2 * (i*2+1) * (i*2+2)) 
    else:
      a -= 4/(i*2 * (i*2+1) * (i*2+2)) 
  pi = a

  def is_even(self,num):
    return True if num%2 == 0 else False
  
  def fac(self,num):
    return math.factorial(num)
  
  def is_prime(self,num):
    for i in range(2,int((num**0.5))+1):
      if num%i == 0: return False
    return True

  def divide_by(self,num,k):
    return True if num%k==0 else False
  
  def ten_to_bi(self,num):
    return bin(num)[2:]
  
  def ten_to_oct(self,num):
    return oct(num)[2:]

  def ten_to_sixteen(self,num):
    return hex(num)[2:].upper()

  def int_to_roman(self,num):
    roman_list = [
        ['','I','II','III','IV','V','VI','VII','VIII','IX'],
        ['','X','XX','XXX','XL','L','LX','LXX','LXXX','XC'],
        ['','C','CC','CCC','CD','D','DC','DCC','DCCC','CM']
    ]
    roman_result = ''
    l = len(str(num))
    for i in range(l):
      # print(i,l-i-1,int(str(num)[i]),roman_list[l-i-1][int(str(num)[i])])
      roman_result += roman_list[l-i-1][int(str(num)[i])]
    return roman_result

# do not submit this code!!!
my_math = MyMath()
num = int(input("Please Enter Number : "))
k = int(input("Divide by? : "))

if my_math.is_even(num):
    print('This number is even number.')
else:
    print('This number is not even number.')

if num <= 20:
    print(f'factorial is {my_math.fac(num):,.0f}.')
else:
    print('factorial TOO LONG')

if my_math.is_prime(num):
    print('This number is a prime number.')
else:
    print('This number is not a prime number.')

if(my_math.divide_by(num,k)):
    print(f'{num} is divisible by {k}')
else:
    print(f'{num} is not divisible by {k}')

print(f'{num} is {my_math.ten_to_bi(num)} in base 2.')
print(f'{num} is {my_math.ten_to_oct(num)} in base 8.')
print(f'{num} is {my_math.ten_to_sixteen(num)} in base 16.')
print(f'{num} is {my_math.int_to_roman(num)} in roman numeral system.')
print(f'PI is {my_math.pi:.20f}')
````

````python
import numpy as np
class Matrix:
  def __init__(self,m):
    self.m = m
  
  def det(self):
    return int(round(np.linalg.det(self.m)))
  
  def show(self):
    for i in self.m:
      for j in i:
        print(f'{j:^6}', end = ' ')
      print('')
  
  def plus(self,B):
    return Matrix(np.add(self.m, B.m))

  def minus(self,B):
    return Matrix(np.subtract(self.m, B.m))

  def multiply(self,B):
    return Matrix(np.matmul(self.m,B.m))

# do not submit this code!!!
def input_matrix():
    data = []
    for i in range(3):
        data.append([int(j) for j in input(f"Row {i+1} : ").split(' ')])
    return data

print("input row of matrix A")
A = Matrix(input_matrix())
print("input row of matrix B")
B = Matrix(input_matrix())

print(f'Det of Matrix(A) = {A.det()}')
print(f'Det of Matrix(B) = {B.det()}')

print("Matrix(A+B) is:")
res = A.plus(B)
res.show()

print("Matrix(A-B) is:")
res = A.minus(B)
res.show()

print("Matrix(A*B) is:")
res = A.multiply(B)
res.show()
````

ให้นิสิตพัฒนาโปรแกรมเพื่อหาว่า string ของวงเล็บที่รับเข้ามามีวงเล็บครบคู่หรือไม่ โดยวงเล็บต้องเรียงให้ถูกลำดับด้วย.

กำหนดให้สร้าง class py_solution และฟังก์ชัน is_valid_parentheses(self) ที่ return ค่าเป็น True เมื่อวงเล็บถูกต้อง และ return ค่าเป็น False เมื่อวงเล็บไม่ถูกต้อง และให้แสดงผลลัพธ์ตามที่กำหนดให้

example
input: ()[]{}
valid parentheses
example
input: {[()]}
valid parentheses
example
input: ([)]
invalid parentheses
example
input: (
invalid parentheses



````python
class py_solution:
  def __init__(self,s):
    self.s = s
  def is_valid_parentheses(self):
    open = ['(','[','{']
    close = [')',']','}']
    stack = []
    for i in self.s:
      # print(stack,i)
      if i in open:
        stack.append(i)
      elif open.index(stack[-1]) == close.index(i):
        stack.pop()
      else: return False
    if stack:
      return False
    return True

sol = py_solution(input("input: "))
print('valid parentheses') if sol.is_valid_parentheses() else print('invalid parentheses')
````

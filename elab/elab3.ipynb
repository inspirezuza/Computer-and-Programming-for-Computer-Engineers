Requirements
Your program must take two inputs: (1) the initial amount in Baht, and (2) the annual interest rate in percent. Both numbers can be any non-negative numbers that can contain fraction.

## 04 Cuboid and its Double
Cuboid
Write a program that reads width, length, and height of cuboid from user. Then, the program finds surface area and volume of this cuboid with 3 decimal points.

After that, the program will double width, length, and height and re-compute surface area and volume of this cuboid with 3 decimal points one more time.

Requirements
You MUST define 3 functions:

compute_rectangle_area(first_side, second_side): receives two sides of rectangle and returns area of rectangle

compute_surface_area(width,length,height): receives width, length, and height of cuboid and returns surface area of cuboid. Make use of compute_rectangle_area in this function

compute_volume(width,length,height): receives width, length, and height of cuboid and return the volume of cuboid
```
Example (shell mode)
>>> compute_rectangle_area(3.5,4.12)
14.42
>>> compute_surface_area(2,3.5,4.12)
59.32000000000001
>>> compute_volume(2,3.5,4.12)

Example
Enter width: 2
Enter height: 3.5

For [2.00 x 3.50 x 4.12] cuboid:

For [4.00 x 7.00 x 8.24] cuboid:

Example
Enter width: 5
Enter length: 10
Enter height: 15
For [5.00 x 10.00 x 15.00] cuboid:

For [10.00 x 20.00 x 30.00] cuboid:
```

````python
def compute_rectangle_area(first_side, second_side):
    return first_side*second_side

def compute_surface_area(width, length, height):
    return 2*(compute_rectangle_area(width, length) + compute_rectangle_area(width, height) + compute_rectangle_area(length, height))

def compute_volume(width, length, height):
    return width*length*height

width = float(input("Enter width: "))
length = float(input("Enter length: "))
height = float(input("Enter height: "))
print(f"For [{width:.2f} x {length:.2f} x {height:.2f}] cuboid:")
print(f"Surface area = {compute_surface_area(width, length, height):.3f}")
print(f"Volume = {compute_volume(width, length, height):.3f}")
print("")
print("After doubling each side,")
width*=2
length*=2
height*=2
print(f"For [{width:.2f} x {length:.2f} x {height:.2f}] cuboid:")
print(f"Surface area = {compute_surface_area(width, length, height):.3f}")
print(f"Volume = {compute_volume(width, length, height):.3f}")
````

````python
import math

def f(x):
    if x < 0: return math.sqrt(x*x+1)
    elif x==0: return x
    elif 0<x<=99: return 3*x*x - (1-x)**2
    else: return 2*x*x*x - x/(math.sqrt(x+1))

x = float(input("Enter x : "))
print(f"f({x:.2f}) = {math.ceil(f(x))}")
````

````python
class Account:
    def __init__(self, name, num, money, history):
        self.name = name
        self.num = num
        self.money = int(money)
        self.history = history

    def show_data(self):
        print(f'Name : {self.name}')
        print(f'Account number : {self.num}')
        print(f'Money : {self.money}')
        print(f'History : {self.history}')

    def update_history(self, payload):
        if self.history == '':
            self.history += payload
        else:
            self.history += '\n' + payload

    def deposit(self):
        amount = int(input('Amount : '))
        self.money += amount
        payload = f'Deposit : {amount}'
        self.update_history(payload)
        print(f'Current money : {self.money}')

    def withdraw(self):
        amount = int(input('amount : '))
        self.money -= amount
        payload = f'Withdraw : {amount}'
        self.update_history(payload)
        print(f'Current money : {self.money}')

    def show_history(self):
        print(self.history)

import json

filename = input('Filename : ')
# filename = 'data1.json'
data = json.load(open(filename))
num = input('Account number : ')
# num = '125463'

for i in data:
    if i['account number'] == num:
        customer = Account(i['name'], i['account number'], i['money'], i['history'])
        break

while 1:
    menu = input('Menu : ')
    if menu == '1':
        customer.show_data()
    elif menu == '2':
        customer.deposit()
    elif menu == '3':
        customer.withdraw()
    elif menu == '4':
        customer.show_history()
    elif menu == '5':
        num = input('Account number : ')
        for i in data:
            if i['account number'] == num:
                customer = Account(i['name'], i['account number'], i['money'], i['history'])
                break
    elif menu == '0':
        break
````

คำอธิบายช่องของข้อมูล

    "_range": ช่วงของระยะทาง (สัมพันธ์กับความเร็ว) มีหน่วยเป็น m
    "velocity": ความเร็วในแต่ละช่วง (มีหน่วยเป็น m/s)
    "response": ค่าการตอบสนอง
คำอธิบาย _range และ velocity ของ A (ตัวอย่างด้านบน)
ช่วงของระยะทางที่วิ่ง (เมตร)	ความเร็วที่วิ่งได้ในช่วงดังกล่าว (เมตร/วินาที)

จะเห็นว่า A และ D ใช้เวลาเท่ากัน แต่ D มี response ที่ดีกว่า A

ดังนั้น D จึงได้ลำดับ 1

คำอธิบาย Example 2
เวลาเหมือน example 1 ทุกประการ แต่ A และ D ก็ยังมี response ที่เท่ากันอีก ดังนั้น A จึงได้ลำดับ 1 เพราะ A นำมาก่อน D ตามพจนานุกรม

````python
import json
filename = input('Enter json filename : ')
data = json.loads(open(filename).read())
distance = int(input('Enter distance (meter) : '))

racers = []

for racer in data["racer"]:
    time = 0
    remain = distance
    if distance > racer['_range'][-1]:
        time += (distance - racer['_range'][-1]) / racer['velocity'][-1]
        remain = racer['_range'][-1]
    for i in range(len(racer['_range'])-1):
        if racer['_range'][i] <= remain <= racer['_range'][i+1]:
            time += (remain - racer['_range'][i]) / racer['velocity'][i]
        else:
            time += (racer['_range'][i+1] - racer['_range'][i]) / racer['velocity'][i]
            
    racers.append([float(f'{time:.2f}'),int(racer['response']),racer['name']])

racers = sorted(racers,key=lambda x: (x[0],-x[1],x[2]))
for i in racers:
    print(i[2], f'{i[0]:.2f}')
````

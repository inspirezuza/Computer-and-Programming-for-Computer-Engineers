#05 Simple Scores
Simple Scores
Write program to read names and scores of students. The program stores names and scores in a dictionary. The program continues reading until user enters 'Q'.

After that the program shows the resulting dictionary, and score that each student got. At the end, the program reports average scores from students who got scores below 60.

Example 1
Enter name: Evan
Enter scores: 52
Enter name: Molly
Enter scores: 54
Enter name: Q
{'Evan': 52.0, 'Molly': 54.0}
Evan got 52.00
Molly got 54.00
Average scores of students who got below 60 = 53.00
Example 2
Enter name: Pat
Enter scores: 94
Enter name: Zoe
Enter scores: 60
Enter name: Ken
Enter scores: 71
Enter name: Q
{'Pat': 94.0, 'Zoe': 60.0, 'Ken': 71.0}
Pat got 94.00
Zoe got 60.00
Ken got 71.00
No one got below 60.
Example 3
Output is

Enter name: Ann
Enter scores: 75
Enter name: Beth
Enter scores: 55
Enter name: Cole

Enter scores: 50

```python
d = {}
below = []
while 1:
    name = input('Enter name: ')
    if name == 'Q': break
    score = float(input('Enter scores: '))
    if score<60:
        below.append(score)
    d[name] = score
print(d)
# print(below)
for name,score in d.items():
    print(f'{name} got {score:.2f}')
if below == []:
    print('No one got below 60.')
else:
    print(f'Average scores of students who got below 60 = {sum(below)/len(below):.2f}')
```

##08 Cloth Size
Cloth Size
สมมติว่าเจ้าของบริษัทต้องการตัดเสื้อให้พนักงานทุกคน จึงต้องการทราบว่า ควรซื้อเสื้อแต่ละขนาดเป็นจำนวนกี่ตัว จงเขียนฟังก์ชัน cloth_size(width_list) เพื่อช่วยเจ้าของบริษัทคำนวณ โดยฟังก์ชันนี้รับลิสต์ width_list ที่เก็บรายการขนาดรอบอกของพนักงานทุกคน และคืนค่าเป็น dictionary ที่เก็บจํานวนเสื้อแต่ละขนาด โดยรอบอกที่เล็กกว่าหรือเท่ากับ 36 จะเป็นขนาด S รอบอกที่ใหญ่กว่า 36 แต่ไม่เกิน 44 จะเป็นขนาด M และรอบอกที่มากกว่า 44 จะเป็นขนาด L

หมายเหตุ ห้ามใช้คำสั่ง count ในไพตอน

ตัวอย่างการทำงาน
 
>>> cloth_size([50, 44, 40, 48])
{'S': 0, 'M': 2, 'L': 2}



```python
def cloth_size(width_list):
    d = {'S':0,'M':0,'L':0}
    for i in range(len(width_list)):
        if width_list[i] <= 36:
            d['S'] += 1
        elif width_list[i] <= 44:
            d['M'] += 1
        else: d['L'] += 1
    return d
```

##09 Count English Characters
Count English Characters
จงเขียนฟังก์ชัน count_char(word) ซึ่งรับสตริง word เข้ามา แล้วสร้าง dictionary เพื่อนับจํานวนครั้งที่ตัวอักขระปรากฏในสตริง word

โดยจะนับทั้งอักขระภาษาอังกฤษตัวเล็กและตัวใหญ่รวมกัน แต่ใน dictionary จะใช้เป็นอักขระภาษาอังกฤษตัวเล็กในการนับ

และไม่สนใจอักขระอื่นใด เช่น ช่องว่าง หรือเครื่องหมาย

หมายเหตุ ห้ามใช้คำสั่ง count

ให้ส่งคำตอบเฉพาะ definition ของ function count_char(word)

ตัวอย่างผลลัพธ์เมื่อเรียกฟังก์ชัน count_char
 
>>> print(count_char('Hello, There'))
{'h': 2, 'e': 3, 'l': 2, 'o': 1, 't': 1, 'r': 1}

```python
import string
def count_char(word):
    d = {}
    for char in word.lower():
        if char not in string.ascii_lowercase:continue
        if char in d:
            d[char] += 1
        else:
            d[char] = 1
    return d
```

##10 นับคำ
นับคำ
จงเขียนฟังก์ชัน count_words(s) ซึ่งรับสตริงของคำที่ขั้นด้วยช่วงว่างมาแล้วสร้าง dictionary ที่จับคู่คำกับจำนวนครั้งที่มันปรากฏในสตริง s

ตัวอย่าง
>>> count_words('hello there')

>>> count_words('spam spam spam bacon and spam')
{'spam': 4, 'bacon': 1, 'and' 1}
>>> count_words('ft qzrr qzrr mya qzrr ft krrw mya krrw krrw gsc krrw ft gsc qzrr krrw')
{'mya': 1, 'gsc': 1, 'krrw': 4, 'ft': 2, 'qzrr': 3}

```python
def count_words(s):
    d = {}
    for word in s.split(' '):
        if word in d:
            d[word] += 1
        else:
            d[word] = 1
    return d
```

##11 การประมูล
การประมูล
ให้นิสิตเขียนโปรแกรมเพื่อเก็บข้อมูลการประมูลสินค้า โดยจะมีการใส่ชื่อเว้นวรรคราคาประมูลซึ่ง 1 คนสามารถประมูลได้หลายครั้ง จนกว่าผู้ใช้จะใส่คำว่า end หลังจากนั้นให้แสดงชื่อผู้ประมูล ราคาประมูลสูงสุดและจำนวนครั้งในการประมูลของคนนั้นๆ เรียงตามลำดับตัวอักษรของชื่อผู้ประมูล และสรุปผลการประมูลว่าใครเป็นผู้ชนะการประมูล ถ้าประมูลราคาเท่ากันให้ผู้ที่ประมูลก่อนชนะ
ตัวอย่าง
ข้อมูลเข้า	ข้อมูลออก
kong 100
saac 500
kong 300
pp 1000
saac 900
end
kong bid at the price of 300.0 baht in 2 times.
pp bid at the price of 1000.0 baht in 1 time.
saac bid at the price of 900.0 baht in 2 times.
The winner is pp.


```python
d = {}
winner = ''
bid = 0
while 1:
    a = input()
    if a == 'end':break
    name,value = a.split(' ')
    if name in d:
        d[name].append(int(value))
    else:
        d[name] = [int(value)]
    
    if int(value) > bid:
        bid = int(value)
        winner = name

for name in sorted(d):
    if len(d[name]) == 1:
        print(f'{name} bid at the price of {max(d[name]):.1f} baht in {len(d[name])} time.')
    else:
        print(f'{name} bid at the price of {max(d[name]):.1f} baht in {len(d[name])} times.')
print(f'The winner is {winner}.')
```

##12 แทนค่าตัวแปร
แทนค่าตัวแปร
จงเขียนโค้ดเพื่อช่วยแทนค่าตัวแปรในโปรแกรมจำลอง โดยเริ่มต้นต้องกำหนดค่าให้กับตัวแปรก่อน หลังจากนั้น จึงสร้างโปรแกรมจำลองที่ใช้ตัวแปรเหล่านั้น โดยตัวแปรที่ใช้ในโปรแกรมจะมีเครื่องหมายปีกกาครอบดังตัวอย่างด้านล่าง โค้ดไพตอนของนิสิตจะต้องแทนค่าตัวแปรในโปรแกรมจำลอง และพิมพ์ผลลัพธ์ออกมา

ให้ใช้ -1 เป็นการระบุว่า ได้กำหนดค่าตัวแปรและสร้างโปรแกรมจำลองเสร็จสิ้น

ตัวอย่างผลลัพธ์ที่ 1
Enter variables and values:
x = 5
y = 6
-1
Enter your program:
z = {x} + {y}
print({x})
print({y})
-1
Result:
z = 5 + 6
print(5)
print(6)

```python
print('Enter variables and values:')
vv = {}
while 1:
    a = input()
    if a == '-1': break
    var, val = [x.strip() for x in a.split('=')]
    vv[var] = val

print('Enter your program:')
program = []
while 1:
    a = input()
    if a == '-1': break
    program.append(a)

print('Result:')
skip = 0
for s in program:
    result = ''
    for i in range(len(s)):
        if skip:
            if s[i] == '}':
                skip = 0
            continue    
        if s[i] == '{':
            skip = 1    
            result += vv[s[i+1:s[i+1:].find('}')+i+1]]
        else:
            result += s[i]
    print(result)
```

    1.กระบวนการถอดรหัสพันธุกรรม (Transcription) หมายถึง การสร้าง RNA โดยใช้ DNA เป็นแม่พิมพ์ โดยวิธีการนี้ ข้อความพันธุกรรมบน DNA จะถูกถ่ายทอดไปยัง RNA
    ลําดับเบสมีทั้งหมด 5 ชนิด ได้แก่ Adenine(A), Cytosine(C), Guanine(G), Thymine(T) และ Uracil(U) โดยสามารถพบ A C G T ได้ใน DNA และ A C G U ได้ใน RNA
        ในกระบวนการนี้ เราจะนํา DNA แม่แบบมาสร้างเป็น RNA โดยจะใช้คู่เบสที่พอดีกับคู่เบสของ DNA แม่แบบตามตารางข้างล่าง ดังนั้น หาก DNA แม่แบบเป็น TACATTAAAGGCCC ก็จะได้ RNA เป็น AUGUAAUUUCCGGG

DNA แม่แบบ        RNA
         A                    U
         C                    G
         G                    C
         T                    A

    2.กระบวนการแปลรหัส (Translation) หมายถึงการแปลรหัสข้อมูลพันธุกรรมซึ่งอยู่ในรูปของ RNA ไปเป็นลําดับกรดอะมิโนในกระบวนการสังเคราะห์โปรตีน
        สําหรับกระบวนการนี้จะเริ่มต้นทํางานที่ลําดับเบส AUG (start codon)ซึ่งจะได้กรดอะมิโน methionine และจะสังเคราะห์กรดอะมิโนรอบละ 3 ลําดับเบส และหยุดการสร้างเมื่อพบลําดับเบส
    UAA UGA หรือ UAG (stop codon) ซึ่งลําดับเบสทั้ง 3 แบบนี้จะไม่ให้กรดอะมิโนออกมา ดังนั้น หาก RNA เป็น GCAUGAGCCAGUACACCUAAAGU ก็จะได้กรดอะมิโนทั้งหมด 5 ตัวคือ
    GC[AUG][AGC][CAG][UAC][ACC]UAAAUG
จากสาย DNA ที่กําหนดให้ จงพัฒนาโปรแกรมเพื่อหาจํานวนกรดอะมิโนที่ได้หลังจากผ่านกระบวนการถอดรหัสพันธุกรรม และ กระบวนการแปลรหัสแล้ว รับประกันว่า DNA ที่นําเข้าจะประกอบด้วย ‘A’, ‘C’, ‘G’ และ ‘T’ เท่านั้น และภายใน DNA จะพบ start codon ก่อน stop codon เสมอ และจะถอดเฉพาะสายรหัสแรกสุดที่เจอเท่านั้น

Sample Output 1
DNA: CGTCGTCCTACAGCATGAGCATCATCAGCCCATTGTCATGC
4 Amino acid(s)
Note: รหัส RNA ที่ได้ GCAGCAGG[AUG][UCG][UAC][UCG]UAGUAGUCGGGUAACAGUACG
Sample Output 2
DNA: CGTACAACGCGGACAAGCGTACCTGCATCTACAAA
8 Amino acid(s)
Note: รหัส RNA ที่ได้ GC[AUG][UUG][CGC][CUG][UUC][GCA][UGG][ACG]UAGAUGUUU

```python
dna2rna = {'A':'U','C':'G','G':'C','T':'A'}
dna = input("DNA: ")
rna = ''
for n in dna:
    rna += dna2rna[n]
start = 'AUG'
stop = ['UAA','UGA','UAG']
head = rna.find(start)
tail = 999999
for j in range(head,len(rna),3):
    if rna[j:j+3] in stop:
        tail = j
        break
# print(rna)
# print(head,tail)
# print(rna[head:tail])
print(f'{(tail-head)//3} Amino acid(s)')
```

```python
class flight:
    def __init__(self,source,destination,amount):
        self.source = source
        self.destination = destination
        self.amount = amount

def printmenu():
    print('''\
Select menu :
1. add flight data
2. flight data by code
3. show all flight data
4. flight booking
5. show people flight data
6. exit\
''')

allflight = {}
def addflight():
    data = input('Add data : ')
    code = data[:7]
    source = data[8:12]
    destination = data[16:20]
    amount = [int(x.strip()) for x in data[21:].split('/')]
    allflight[code] = flight(source,destination,amount)

def code2flight(code):
    source = allflight[code].source
    destination = allflight[code].destination
    amount = allflight[code].amount
    print(f'{code} is from {source} to {destination}, number of people booking is {amount[0]}/{amount[1]}')

def showall():
  print('All flight data')
  for code in allflight:
    code2flight(code)

book = {}
def booking():
  name = input('Enter your name : ')
  code = input('Enter flight code : ')
  amount = allflight[code].amount
  if amount[0] >= amount[1]:
    print('The flight is full, Sorry')
  else:
    allflight[code].amount[0] += 1
    if name in book:
      book[name].append(code)
    else:
      book[name] = [code]
    print('Success')

def name2code():
  name = input('Enter your name : ')
  result = ''
  for code in book[name]:
    result += code + ' '
  print(f'{name} is booking flight code = {result}')

printmenu()
while 1:
    menu = input('menu : ')
    if menu == '1':
        addflight()
    elif menu == '2':
      code = input('Enter code : ')
      code2flight(code)
    elif menu == '3':
        showall()
    elif menu == '4':
        booking()
    elif menu == '5':
        name2code()
    elif menu == '6':
        print('EXIT!!!!!!!!!!!!!!!')
        break
```

#19 L13 sibling
sibling
ข้อความเข้ามา N ข้อความ โดยจะมาในรุปแบบ

... is xxx of ... โดยที่ xxx คือ ความสัมพันธ์

หลังจากนั้น จะมีคำถามว่า ... และ ... เป็นญาติกันหรือไม่ โดยถ้าเป็นญาติกันจะต้องมี พ่อหรือแม่ เป็นคนเดียวกัน

Example 3

C is father of B
C is father of X
C is wife of Z
is X and B are sibling

```python
class Sibling:
    def __init__(self,father=None,mother=None):
        self.father = father
        self.mother = mother
    def __str__(self):
        return f'{self.father}, {self.mother}'

people = {}

def name(line):
    return (line[:line.find(' ')],line[-1::-1][:line[-1::-1].find(' ')])
n = int(input())
for i in range(n):
    line = input()
    if 'mother' in line[5:]:
        a,b = name(line)
        people[b] = Sibling(mother=a)
    elif 'father' in line[5:]:
        a,b = name(line)
        people[b] = Sibling(father=a)

a,b = input().split(' and ')
a = a[-1::-1][:a[-1::-1].find(' ')]
b = b[:b.find(' ')]

if a not in people or b not in people:
    print('No')
elif people[a].mother == people[b].mother:
    print('Yes')
elif people[a].father == people[b].father:
    print('Yes')
else: print('No')
```

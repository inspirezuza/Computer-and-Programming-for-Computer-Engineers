##02 Score Statistics from File
Score Statistics from File
Write a Python program that reads a sequence of scores from the specified file. The program then reports the following:

All the scores and their indexes, starting from 1,
The average of the scores with two decimal places,
The minimum score, and
The maximum score
The program must be able to correctly ignore empty lines in the input file.
```
Examples
Suppose the file scores1.txt contains the following:

10
13
19
The output of the program when specifying scores1.txt as input is:

Enter score file: scores1.txt
Student #1 score: 10
Student #2 score: 13

Average score is 14.00
Minimum score is 10
Maximum score is 19

Python Program
```

````python
score_file = input(f'Enter score file: ') 
scores = [int(x) for x in open(f'{score_file}').read().splitlines() if x != '']
for i in range(len(scores)):
  print(f'Student #{i+1} score: {scores[i]}')
print(f'Average score is {sum(scores)/len(scores):.2f}')
print(f'Minimum score is {min(scores)}')
print(f'Maximum score is {max(scores)}')
````

##03 GPA Calculator
GPA Calculator
Write a Python program that reads a CSV file containing a list of subjects, their credits, and the grades received. The program then prints out the grade summary, the total credits, and the grade point average (GPA) with two decimal places.

Each line in a grade data file is of this form:

subject-code,credits,grade
The program must be able to correctly ignore empty lines in the input file.

##04 Oldest People
Oldest People
Write a Python program that reads a CSV file containing a list of people and their ages. The program then reports the name(s) of the oldest person(s) in the list.

Notes: if there is more than one person with the same oldest age, the program must print out these names in the same order as listed in the input file.
```
Examples
Suppose the file age1.txt contains the following:

Somchai,58
Somying,50
Somsak,62
The output of the program when specifying age1.txt as input is:

Enter age file: age1.txt
Oldest person(s) with the age of 62:
- Somsak
In case there is more than one person with the oldest age, the program must report all those people's names. Suppose the file age2.txt contains the following:

Sophia,95
Jackson,56
Emma,95
Aiden,42
Olivia,68
Lucas,75
Ava,86
Liam,92
The output of the program when specifying age2.txt as input is:

Enter age file: age2.txt
Oldest person(s) with the age of 95:
- Sophia
- Emma
```

````python
filename = input('Enter age file: ')
# filename = 'g.txt'
lines = open(filename).read().splitlines()
table = [x.split(',') for x in lines]
max = -9999
oldest_name = []
for i in range(len(table)):
    if int(table[i][1]) > max:
        oldest_name = []
        max = int(table[i][1])
        oldest_name.append(table[i][0])
    elif int(table[i][1]) == max:
        oldest_name.append(table[i][0])

print(f'Oldest person(s) with the age of {max}:')
for i in oldest_name:
    print(f'- {i}')
````

##09 L12 Matrix
Matrix
ให้นิสิต CPE34 ออกแบบโปรแกรมแสดงเมตริกซ์ผลลัพธ์จากไฟล์ โดยรับ input เป็นชื่อไฟล์

```
ให้แสดงผล matrix ด้วย format นี้
for i in range(len(res)):
    for j in range(len(res[0])):
        print(f"{res[i][j]:^5}",end=" ")
    print()
example
File name: matrix.txt

 485   550   616
```

````python
def printmat(res):
    for i in range(len(res)):
        for j in range(len(res[0])):
            print(f"{res[i][j]:^5}",end=" ")
        print()

def mulmat(m1,m2):
    mat = []
    for i in range(len(m1)):
        t = []
        for j in range(len(m2[0])):
            sum = 0
            for k in range(len(m1[0])):
                sum += m1[i][k]*m2[k][j]
            t.append(sum)
        mat.append(t)
    return mat

def plusmat(m1,m2):
    mat = []
    for i in range(len(m1)):
        t = []
        for j in range(len(m1[0])):
            t.append(m1[i][j] + m2[i][j])
        mat.append(t)
    return mat

def minusmat(m1,m2):
    mat = []
    for i in range(len(m1)):
        t = []
        for j in range(len(m1[0])):
            t.append(m1[i][j] - m2[i][j])
        mat.append(t)
    return mat

filename = input('File name: ')
# filename = 'g.txt'
lines = open(filename).read().splitlines()
table = [x.split(' ') for x in lines]

operand = []
mat = []
t = []
for i in range(len(table)):
    if table[i][0] in ['+','-','*']:
        mat.append(t)
        t = []
        operand.append(table[i][0])
    else:
        t.append([int(x) for x in table[i]])
mat.append(t)


for i in range(len(operand)):
    if operand[i] == '*':
        ans = mulmat(mat[i],mat[i+1])
    elif operand[i] == '+':
        ans = plusmat(mat[i],mat[i+1])
    elif operand[i] == '-':
        ans = minusmat(mat[i],mat[i+1])
    mat[i+1] = ans
printmat(ans)
````

##14 L12 GcdLcm
[GCD LCM] ห.ร.ม. ค.ร.น.
ให้นิสิตเขียนโปรแกรมเพื่อหาค่า ห.ร.ม. และ ค.ร.น. ของจำนวนเต็มบวกทั้ง 2 จำนวน
```
Example 1
a : 8
b : 30
GCD : 2
LCM : 120
Example 2
a : 6
b : 12
GCD : 6
LCM : 12
Example 3
a : 8
b : 30
GCD : 2
LCM : 120
```

````python
a = int(input('a : '))
b = int(input('b : '))

def gcd(a,b):
    for i in range(min(a,b),0,-1):
        if a%i == 0 and b%i == 0:
            return i

def lcm(a,b):
    for i in range(max(a,b),a*b+1):
        if i%a == 0 and i%b == 0:
            return i

print(f'GCD : {gcd(a,b)}')
print(f'LCM : {lcm(a,b)}')
````

##16 L12 word_chain
Word chain โซ่คำ
โซ่คำ คือลำดับของคำที่มีจำนวนอักขระเท่ากันและแต่ละคำที่มีลำดับติดกันจะต้องมีตำแหน่งที่มีตัวอักขระต่างกันไม่เกินสองตำแหน่ง เช่น
```
    HEAD และ HEAP จำต่างกันตำแหน่งเดียว คือ D และ P ในตำแหน่งตัวอักขระที่ 4 ของคำ
ตัวอย่างที่ไม่เป็นโซ่คำ เช่น
    REAR และ BAER จะต่างกัน 3 ตำแหน่ง คือ ตำแหน่งที่ 1 (R กับ B) ตำแหน่งที่ 2 (E กับ A) และ ตำแหน่งที่ 3 (A กับ E)

ตัวอย่างของโซ่คำที่ต่อเนื่อง ได้แก่ HEAD HEAP LEAP TEAR REAR และ EGG EAG GAE GAP TAP TIN
ตัวอย่างของโซ่คำที่ขาด ได้แก่ LEAP TEAR REAR BAER BAET BEEP ซึ่งจะขาดที่คำว่า RAER และ BAER ได้เป็นโซ่ 2 เส้น
```
จงเขียนโปรแกรมเพื่อแสดงว่ามีโซ่คำสั้นทั้งหมดกี่เส้นในข้อความ และ เส้นที่ยาวที่สุดยาวเท่าไร รับประกันว่าแต่ละคำจะมีจำนวนตัวอักษรเท่ากัน และ ประกอบด้วยตัวอักษรภาษาอังกฤษพิมพ์ใหญ่ ('A' ถึง 'Z') เท่านั้น
```
Example 1
Text: HEAD HEAP LEAP TEAR REAR BAER BAET BEEP JEEP JOIP JEIP AEIO
2 Chain(s). Maximum length is 7 word(s).
Note: สามารถแบ่งได้ 2 เส้น ดังนี้
HEAD HEAP LEAP TEAR REAR มีความยาวเป็น 5 คำ
BAER BAET BEEP JEEP JOIP JEIP AEIO มีความยาวเป็น 7 คำ
Example 2
Text: SUN TON BOW GOD LOT KID FAX BAT FAT CAR EAT FEE SEA MAP DRY SPY TAP
7 Chain(s). Maximum length is 5 word(s).
Note: สามารถแบ่งได้ 2 เส้น ดังนี้
SUN TON BOW GOD LOT มีความยาวเป็น 5 คำ
KID มีความยาวเป็น 1 คำ
FAX BAT FAT CAR EAT มีความยาวเป็น 5 คำ
FEE SEA มีความยาวเป็น 2 คำ
MAP มีความยาวเป็น 1 คำ
DRY SPY มีความยาวเป็น 2 คำ
TAP มีความยาวเป็น 1 คำ
```

````python
textinput = input('Text: ')
# textinput ='SUN TON BOW GOD LOT KID FAX BAT FAT CAR EAT FEE SEA MAP DRY SPY TAP'
text = [x for x in textinput.split(' ')]

chain = 1
maxstreak = -999
streak = 1
for i in range(1,len(text)):
    # print(i,text[i],end = ' ')
    diff = 0
    for j in range(len(text[0])):
        if text[i-1][j] != text[i][j]:
            # print(text[i-1][j], text[i][j], end=' | ')
            diff += 1
            if diff>2: break
    # print('')
    if diff > 2:
        chain += 1
        # print(chain,streak)
        diff = 0
        if streak > maxstreak:
            maxstreak = streak
        streak = 1
    else:
        streak += 1

if streak > maxstreak:
    maxstreak = streak

print(f'{chain} Chain(s). Maximum length is {maxstreak} word(s).')
````

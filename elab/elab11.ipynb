Radio

String mode: the current mode which is "FM" or "AM"
double frequency: the current frequency of radio. The range of frequency depends on the radio mode.
FM: [87.5 MHz , 108 MHz]
AM: [150 kHz , 280 kHz]
Here are the list of radio's behaviours. Note that the mutator methods have more detailed than the commonly used ones. Make sure you adjust the mutator methods as specified.

Radio()	Construct of radio object. the default radio mode is "FM"
void	set_mode(mode)	Change the radio mode which is either "FM" or "AM", and set the frequency of radio to be the minimum frequency allowed of such mode.	
void	set_frequency(frequency)	Set the current frequency to be the input frequency.
The result frequency must be in the range of current radio mode.
If the input frequency is out of range, the current frequency will not be changed.
boolean	adjust_frequency(frequency)	Adjust the frequency from current frequency to current frequency+x
x can be any double value
The new frequency will be updated if it is in the radio mode range.
The new frequency will not be updated if it is out of the radio mode range.
Return true if the new frequency is updated. Otherwise, return false.
String	__str__()	return the string that are in this format, e.g. "FM Radio: 87.5 MHz",
note that the number must be formatted as a single-place decimal (###.#)
Don't forget getter & setter method for mode and frequency. ( get_mode, get_frequency )

Assume that the mode will be only "FM" or "AM" and no other string value.

````python
class Radio():
  def __init__(self,Mode='FM',frequency=87.5):
    self.__Mode = Mode
    self.__frequency = frequency

  def get_mode(self):
    return self.__Mode
  def get_frequency(self):
    return self.__frequency

  def set_mode(self,mode):
    self.__Mode = mode
    self.__frequency = 150
    if mode == 'FM':
      self.__frequency = 87.5

  def set_frequency(self,frequency):
    if self.__Mode == 'FM' and 87.5<=frequency<=108:
      self.__frequency = frequency
    elif 150<=frequency<=280:
      self.__frequency = frequency
  
  def adjust_frequency(self,frequency):
    new = self.__frequency + frequency
    if 87.5<=new<=108:
      self.__frequency = new
      self.__Mode = 'FM'
      return True
    elif 150<=new<=280:
      self.__frequency = new
      self.__Mode = 'AM'
      return True
    return False
  
  def __str__(self):
    unit = 'MHz' if self.__Mode == 'FM' else 'kHz'
    return f'{self.__Mode} Radio: {self.__frequency:.1f} {unit}'
````

Submit your answer for only Line class below. Do not include Application or Point class.

````python
class Line:
    def __init__(self, x1, y1, x2, y2):
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2
        self.slope = (y2 - y1) / (x2 - x1)
        self.yintercept = y1 - (self.slope * x1)

    def contains(self, x, y):
        return (y == (self.slope * x) + self.yintercept) and \
               ((x >= self.x1 and x <= self.x2) or (x >= self.x2 and x <= self.x1)) and \
               ((y >= self.y1 and y <= self.y2) or (y >= self.y2 and y <= self.y1))

    def get_distance(self):
        return ((self.x1 - self.x2) ** 2 + (self.y1 - self.y2) ** 2) ** 0.5

    def get_x1(self):
        return self.x1

    def get_y1(self):
        return self.y1

    def get_x2(self):
        return self.x2

    def get_y2(self):
        return self.y2

    def getY(self, x):
        if ((x >= self.x1 and x <= self.x2) or (x >= self.x2 and x <= self.x1)):
            return (self.slope * x) + self.yintercept
        else:
            return -999.999
````

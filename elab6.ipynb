````python
def list_factors(n):
   """
   Get string of factors of n
   :params n is an integer to find factors
   :return string of factors (with a space between each factor)
   >>> list_factors(6)
   '1 2 3 6 '
   >>> list_factors(7)
   '1 7 '
   >>> list_factors(12)
   '1 2 3 4 6 12 '
   """
   
  result = ""
  for i in range(1,n+1):
      if n%i == 0: result += f'{i} '
  return result


def find_sum_and_num_factors(n):
   """
   Find summation and count number of factors of n
   :params n is an integer to find factors
   :return sum is summation of factors of n
           count is number of factors of n
   >>> find_sum_and_num_factors(6)
   (12, 4)
   >>> find_sum_and_num_factors(7)
   (8, 2)
   >>> find_sum_and_num_factors(12)
   (28, 6)
   """
    
  sum = 0
  counter = 0

  for i in range(1,n+1):
      if n%i == 0: 
          sum += i
          counter +=1
  return (sum, counter)


def isPrime(n):
    if n<2: return False
    for i in range(2,n//2+1):
        if n%i == 0: return False
    return True

n = int(input("Enter positive number: "))
factors = list_factors(n)
sum,counter = find_sum_and_num_factors(n)

print(f"Factors of {n} are")
print(factors)
print(f"Sum of {factors}is {sum}")
print(f"Number of factors is {counter}")
if isPrime(n):
  print(f"{n} is prime number.")
else:
  print(f"{n} is not prime number.")
````

      ให้นืสิตพัฒนาโปรแกรมที่รับวันที่ตามปฏิทิน แล้วคำนวณลำดับของวันที่ดังกล่าวในปีนั้นๆ (1 ถึง 366)

   ( ปีอธิกสุรทิน สามารถคำนวณโดยเป็นปีที่หาร 4 ลงตัวแต่หาร 100 ไม่ลงตัว ยกเว้นปีที่หาร 400 ลงตัว )

````python
def is_leap(y):
    return (y % 4 == 0 and y % 100 != 0) or y % 400 == 0

def date(d,m,y):
    month = [31,28,31,30,31,30,31,31,30,31,30,31]
    if is_leap(y): month[1] = 29
    return sum(month[:m-1]) + d

d = int(input("d: "))
m = int(input("m: "))
y = int(input("y: "))

print(date(d,m,y))
````

````python
l = [i for i in input().split(" ")]
while 1:
    command, num = input().split(" ")
    num = int(num)
    if command == "A":
        l.append(num)
    if command == "S":
        print(l[num])
    if command == "R":
        l.pop(num)
    if command == "E":
        for i in l:
            print(i,end=" ")
        break
````

##12 L06 commaCode
CommaCode
เพื่อให้โปรแกรมสามารถทำงานได้ตามตัวอย่าง
ให้นิสิตสร้าง function ชื่อ commaCode(myList) ซึ่งรับ parameter 1ตัวชื่อ myList
โดยที่ function นี้จะ return เป็น string ของ value ที่อยู่ใน myList ทั้งหมด คั่นด้วย comma และช่องว่าง
นอกจากนี้เมื่อถึง value ตัวสุดท้ายของ myList ให้เชื่อมด้วย and แล้วตามด้วย value ตัวสุดท้าย

note: function ต้องสามารถรับค่า list ว่าง [] ได้โดยที่ไม่เกิด error

```
Example 1
Input: apples bananas tofu cats

Example 2
Input: cpe35

>>> 
Example 3
Input: 
>>> 
Examples 1 (shell mode)
>>> myList = ['apples','bananas','tofu','cats']
>>> commaCode(myList)
apples, bananas, tofu, and cats
>>> 
Examples 2 (shell mode)
>>> myList = ['cpe35']
>>> commaCode(myList)

>>> myList = []
>>> commaCode(myList)

```

````python
def commaCode(myList):
    s = ''
    if myList == ['']: return ''
    if len(myList) == 1: return myList[0]
    for idx, i in enumerate(myList):
        if idx == len(myList)-1:
            s+= f'and {i}' 
        else: s += f'{i}, '
    return s

s = [i for i in input("Input: ").split(" ")]
print(commaCode(s))
````
